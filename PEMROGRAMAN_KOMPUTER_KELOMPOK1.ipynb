```
import numpy as np
import pandas as pd

# --- Load Dataset ---
df = pd.read_csv("Oil_and_Gas_Summary_Production_Data__1967-1999.csv")

# --- Convert kolom ke NumPy Array ---
oil = df["Oil Produced, bbl"].to_numpy()
gas = df["Gas Produced, Mcf"].to_numpy()

# --- Statistik Dasar Oil ---
print("=== OIL PRODUCED (bbl) ===")
print(f"Mean   : {np.mean(oil):.2f}")
print(f"Max    : {np.max(oil):.2f}")
print(f"Min    : {np.min(oil):.2f}")
print(f"Std    : {np.std(oil):.2f}")

# --- Statistik Dasar Gas ---
print("\n=== GAS PRODUCED (Mcf) ===")
print(f"Mean   : {np.mean(gas):.2f}")
print(f"Max    : {np.max(gas):.2f}")
print(f"Min    : {np.min(gas):.2f}")
print(f"Std    : {np.std(gas):.2f}")

# --- Perbandingan Total Produksi ---
print("\n=== PERBANDINGAN OIL vs GAS ===")
print(f"Total Oil : {np.sum(oil):,.0f} bbl")
print(f"Total Gas : {np.sum(gas):,.0f} Mcf")
print(f"Korelasi  : {np.corrcoef(oil, gas)[0,1]:.4f}")
```

```
sns.barplot(data=df.head(10000), x='Production Year', y='Inactive Oil Wells' , hue='Production Year')
plt.xticks(rotation=90)
plt.show()


sns.barplot(data=df.head(10000), x='Production Year', y='Active Oil Wells' , hue='Production Year')
plt.xticks(rotation=90)
plt.show()
```

```
sns.histplot(df['Oil Produced, bbl'].replace([np.inf, -np.inf], np.nan).dropna())
plt.title('Oil Production per Year'), plt.xlabel('Oil Production, bb;'), plt.ylabel('Production Year'), plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print(" Simulated Oil Produced in Year(s) ")
df['Gas Production, Mcf'] = np.random.normal(loc=1000, scale=300, size=len(df))
df['Gas Production, Mcf'] = df['Gas Production, Mcf'].apply(lambda x: max(10, x))
df['Production Date Entered'] = np.random.normal(loc=50, scale=20, size=len(df))
df['Production Date Entered'] = df['Production Date Entered'].apply(lambda x: max(1, x))

df['gor'] = df.apply(lambda row: row['Gas Production, Mcf'] / row['Oil Produced, bbl'] if row['Oil Produced, bbl'] > 0 else 0, axis=1)

print("Descriptive statistics for GOR:")
display(df['gor'].describe())
sns.histplot(df['gor'])
plt.title('Distribution of Gas Oil Ratio (GOR)'), plt.xlabel('GOR (Mcf/bbl)'), plt.ylabel('Production Year'), plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print("Top 5 entries by GOR:")
display(df[['Oil Produced, bbl', 'Gas Production, Mcf', 'Production Date Entered', 'gor']].sort_values(by='gor', ascending=False).head())
```

```
sns.boxplot(df['Oil Produced, bbl'].replace([np.inf, -np.inf], np.nan).dropna())
plt.title('Oil Production per Year'), plt.xlabel('Oil Production, bb;'), plt.ylabel('Production Year'), plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print(" Simulated Oil Produced in Year(s) ")
df['Gas Production, Mcf'] = np.random.normal(loc=1000, scale=300, size=len(df))
df['Gas Production, Mcf'] = df['Gas Production, Mcf'].apply(lambda x: max(10, x))
df['Production Date Entered'] = np.random.normal(loc=50, scale=20, size=len(df))
df['Production Date Entered'] = df['Production Date Entered'].apply(lambda x: max(1, x))

df['gor'] = df.apply(lambda row: row['Gas Production, Mcf'] / row['Oil Produced, bbl'] if row['Oil Produced, bbl'] > 0 else 0, axis=1)

print("Descriptive statistics for GOR:")
display(df['gor'].describe())
sns.boxplot(df['gor'])
plt.title('Distribution of Gas Oil Ratio (GOR)'), plt.xlabel('gor'), plt.ylabel('Production Year'), plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print("Top 5 entries by GOR:")
display(df[['Oil Produced, bbl', 'Gas Production, Mcf', 'Production Date Entered', 'gor']].sort_values(by='gor', ascending=False).head())
```
